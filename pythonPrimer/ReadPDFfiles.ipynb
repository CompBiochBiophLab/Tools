```python
import tabula
import os
import pandas as pd

file1 = "https://nbviewer.jupyter.org/github/kuruvasatya/Scraping-Tables-from-PDF/blob/master/data1.pdf"
file2 = os.path.expanduser('~/Downloads/CA.pdf')
table1 = tabula.read_pdf(file2,pages=1)
table2 = tabula.read_pdf(file2,pages=2)
table1[0],table2[0]
df=pd.concat([table1[0],table2[0]])
df.head
```

```python
df.dtypes
```

```python
df.dropna(subset=['Qualificació'], inplace=True)
df[['nota num','nota text','garbage']] = df['Qualificació'].apply(lambda x: pd.Series(str(x).split(" ")))
df['nota num'] = df['nota num'].apply(pd.to_numeric)
df['weighted']=df['Crèdits']*df['nota num']
df.head
```

```python
tot_credits=df['Crèdits'].sum()
tot_grades=df['weighted'].sum()
weighted_sum = tot_grades/tot_credits
print('nota promig grau: ',weighted_sum)
```
