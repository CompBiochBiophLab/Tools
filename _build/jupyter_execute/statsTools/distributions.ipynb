## [Normal distribution](https://www.khanacademy.org/math/statistics-probability/modeling-distributions-of-data/more-on-normal-distributions/v/introduction-to-the-normal-distribution)

Adapted from [geeksforgeeks](https://www.geeksforgeeks.org/how-to-find-probability-distribution-in-python/).

The normal distribution, also called Gaussian distribution, is a symmetric probability distribution centered on the mean.  Here, the function `stats.norm()` generates continuous random data. `scale`  refers to the standard deviation and `loc` to the mean. plt.distplot() is used to visualize the data. 

$$ f(x) = \frac{1}{\sigma\sqrt{2\pi}}
  \exp\left( -\frac{1}{2}\left(\frac{x-\mu}{\sigma}\right)^{\!2}\,\right)$$

```python
# import packages
import scipy.stats as stats
import seaborn as sns
import matplotlib.pyplot as plt


# generate data
data =stats.norm(scale=1, loc=0).rvs(1000)

# plotting a histogram
ax = sns.histplot(data,
				bins=50,
                kde=True,  # a kernel distribution estimation is used to fit a line
				color='red')
ax.set(xlabel='Normal Distribution', ylabel='Frequency')

plt.show()
```

## [Binomial distribution](https://www.khanacademy.org/math/statistics-probability/random-variables-stats-library/binomial-random-variables/v/binomial-distribution)

Under a given set of factors or assumptions, the binomial distribution expresses the likelihood that a variable will take one of two outcomes or independent values. `np.random.binomial()` is used to generate binomial data. `n` refers to a number of trails and `p` refers to the probability of a particular outcome in each trail. 
$$P(X=k)=\binom{N}{k}\cdot p^k q^{N-k}$$

```python
# import packages
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
 
# generate data
# n== number of trials,
# p== probability in each trial
n, p = 10, .6
data = np.random.binomial(n, p, 10000)
 
# plotting a histogram
ax = sns.histplot(data,
                  bins=20,
                  kde=False,
                  color='red')
ax.set(xlabel='Binomial Distribution', ylabel='Frequency')
 
plt.show()
```

## [Poisson distribution](https://www.khanacademy.org/math/statistics-probability/random-variables-stats-library/poisson-distribution/v/poisson-process-1)

It illustrates how many times an event is expected to happen over a certain amount of time. Itâ€™s also called count distribution. `np.random.poisson()`  is the function used to create data for Poisson distribution. `lam` refers to the number of occurrences that are expected to occur in a given time frame.

$$P(X=k)=e^{-\lambda}\frac{\lambda^k}{k!}, \qquad k=0,1,2,...$$

```python
# import packages
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# generate poisson data
poisson_data = np.random.poisson(lam=5, size=1000)

# plotting a histogram
ax = sns.histplot(poisson_data,
				kde=False,
				color='blue')
ax.set(xlabel='Poisson Distribution', ylabel='Frequency')

plt.show()
```
